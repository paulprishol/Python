# Exercise 03. Selects and aggregations

```python
import pandas as pd
```

## Load the JSON file that you created in the previous exercise into a dataframe

```python
df = pd.read_json('../data/auto.json', orient='records')
df.set_index('CarNumber', inplace=True)
df
```

## Make the following selections

```python
df[df['Fines'] > 2100]
```

```python
df[(df['Fines'] > 2100) & (df['Refund'] == 2)]
```

```python
df[df['Model'].isin(['Focus', 'Corolla'])]
```

```python
df[df.index.isin(['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS'])]
```

## Make the aggregations with the `make` and `model`

```python
df.groupby(by='Make').agg({'Fines': ['median']})
```

```python
df.groupby(by=['Make', 'Model']).agg({'Fines': ['median']})
```

```python
df.groupby(by=['Make', 'Model']).agg({'Fines': ['count']})
```

```python
df.groupby(by=['Make', 'Model']).agg({'Fines': ['min', 'max']})
```

```python
df.groupby(by=['Make', 'Model']).agg({'Fines': ['std']})
```

## Make aggregations with car numbers

```python
df.groupby(by=['CarNumber']).agg({'Fines': ['count']}).sort_values(by=('Fines', 'count'), ascending=False)
```

```python
df[df.index == 'Y7689C197RUS']
```

```python
df.groupby(by='CarNumber').agg({'Fines': ['sum']}).sort_values(by=('Fines', 'sum'), ascending=False)
```

```python
df[df.index == 'X758HY197RUS']
```

```python
df.groupby(by=['CarNumber', 'Model']).agg({'Model' : ['nunique']}).sort_values(by=('Model', 'nunique'), ascending=False)
```
