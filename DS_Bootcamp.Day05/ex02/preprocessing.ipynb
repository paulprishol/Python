# Exercise 02. Preprocessing

```python
import pandas as pd
```

## Download and read the CSV file, making `ID` the index column

```python
df = pd.read_csv('../data/auto.csv', index_col="ID")
```

## Count the number of observations using the method `count()`

```python
df.count()
```

## Drop the duplicates, taking into account only the following columns: `CarNumber`, `Make_n_Model`, and `Fines`

```python
df.drop_duplicates(subset=['CarNumber', 'Make_n_model', 'Fines'], keep='last', inplace=True)
df.count()
```

## Work with missing values

```python
df.isnull().sum()
```

```python
df.dropna(axis='columns', thresh=500, inplace=True)
df.isnull().sum()
```

```python
df['Refund'] = df['Refund'].ffill()
df.isnull().sum()
```

```python
mean = df['Fines'].mean()
df['Fines'] = df['Fines'].fillna(value=mean)
df.isnull().sum()
```

## Split and parse the make and model

```python
df['Make'] = df['Make_n_model'].apply(func=lambda x: x.split()[0])
df['Model'] = df['Make_n_model'].apply(func=lambda x: x.split()[1] if len(x.split()) == 2 else None)
df.drop('Make_n_model', axis='columns', inplace=True)
df
```

```python
df.to_json('../data/auto.json', orient='records', double_precision=1)
```
