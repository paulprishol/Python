# Exercise 01. Basic operations

```python
import pandas as pd
```

## Create a dataframe called `views` with two columns: `datetime` and `user` by reading `feed-views.log`

```python
df = pd.read_csv('../data/feed-views.log', names=['datetime', 'user'],
                 parse_dates=['datetime'], engine='python', sep='\t')
df['year'] = df['datetime'].dt.year
df['month'] = df['datetime'].dt.month
df['day'] = df['datetime'].dt.day
df['hour'] = df['datetime'].dt.hour
df['minute'] = df['datetime'].dt.minute
df['second'] = df['datetime'].dt.second
df.info()
df
```

## Create the new column `daytime`

```python
df.set_index('user', inplace=True)
df['daytime'] = pd.cut(df['hour'], bins=[0, 4, 7, 11, 17, 20, 24],
                       labels=['night', 'early morning', 'morning', 'afternoon', 'early evening', 'evening'],
                       right=False)
df
```

## Calculate the number of elements in your dataframe

```python
print(df.count())
df.value_counts('daytime')
```

## Sort the values in your dataframe by hour, minute, and second in ascending order simultaneously, not one by one

```python
df.sort_values(by=['hour', 'minute', 'second'], inplace=True)
df
```

## Calculate the minimum and maximum for the hours and the mode for the daytime categories

```python
df_night = df[df['daytime'] == 'night']
df_morning = df[df['daytime'] == 'morning']

max_night = df_night['hour'].max()
min_morning = df_morning['hour'].min()

print(f'Max night hour: {max_night}')
print(f'Min morning hour: {min_morning}')
```

```python
df[df['hour'] == max_night]
```

```python
df[df['hour'] == min_morning]
```

```python
df['hour'].mode()
```

```python
df['daytime'].mode()
```

## Show the three earliest and latest hours of the day and their corresponding usernames using `nsmallest()` and `nlargest()`

```python
df.nsmallest(3, 'hour')
```

```python
df.nlargest(3, 'hour')
```

## Use the `describe()` method to get the basic statistics for the columns

```python
iqr = df.describe()

print(iqr)

q1 = iqr['hour']['25%']
q3 = iqr['hour']['75%']

print(f'Interquartile range: {q3 - q1}')
```
