### Посмотрим на данные, которые анализирует класс Ratings

### Методы класса Links

Практически вся информация сейчас - это (в том или ином виде) ссылки на другую информацию. Начиная с сайтов и заканчивая ссылками на древнюю литературу в книгах и научных работах. И очень важно уметь грамотно обрабатывать эти ссылки. В нашем случае скурпулёзному аналиу подлежит сайт IMDB - одна из крупнейших площадок с базой данных о фильмах и системой их рецензирования. Давайте же посмотрим, насколько успешно были вытащены оттуда все данные...

p.s. В данном классе было решено поставить ограничение на количество обрабатываемых фильмов из базы данных (links.csv) в максимум 100 строк. Иначе код работал бы непростительно долго, и любой проверяющий скорее пошёл бы смотреть новенькое кино/сериальчик, чем ждать, пока всё это запустится. Так что имейте ввиду и не серчайте.

#### get_imdb()

Данный метод получает на вход список id фильмов и список полей, информацию из которых нужно взять со страницы соответствующего фильма на сайте IMDB. Сначала список фильмов сортируется. После идёт проверка на наличие фильма в базе данных. Далее с помощью requests запрашивается информация со страницы фильма. После обработки возможных ошибок метод проходится по списку полей, и по каждому из них ищет информацию в коде сайта с помощью BeautifulSoup. Вся эта информация заносится в список. В качестве результата возвращается список таких списков. Данный метод служит подспорьем для остальных методов класса.

```python
list_of_movies = [1, 2, 3, 4, 5]
list_of_fields = ['Director', 'Budget', 'Cumulative Worldwide Gross', 'Runtime']
links.get_imdb(list_of_movies, list_of_fields)
```

```python
%timeit links.get_imdb(list_of_movies, list_of_fields)
```

#### top_directors()

Возвращает топ-n режиссёров по количеству срежесированных фильмов. Собирает данные с сайта IMDB с помощью метода get_imdb(). При построчном считывании данных добавляет в словарь пару ключ - значение, где ключом является режиссёр, а значением - единица-счётчик. Если в словаре уже есть соответствующее имя, то к счётчику добавляется 1. После чего словарь сортируется по значениям в порядке убывания.

```python
links.top_directors(10)
```

```python
%timeit links.top_directors(10)
```

#### most_expensive()

Возвращает топ-n самых дорогих в производстве фильмов. Принцип работы аналогичен предыдущему методу за исключением обработки данных о бюджете. Числовая информация о затраченной на фильм сумме извлекается из строки и сохраняется в переменную. Далее (и так для каждой строки данных) заполняется словарь, где ключом является название фильма, а значением - его бюджет в двух форматах: строковом с символом валюты и числовом. В конце происходит сортировка по числовым значениям, на основе которых словарь заполняется заново, но в качестве значений уже сохраняется только строковый вариант представления бюджета.

```python
links.most_expensive(10)
```

```python
%timeit links.most_expensive(10)
```

#### most_profitable()

Возвращает топ-n фильмов по количеству прибыли за вычетом затрат на создание. Принцип работы практически идентичен предыдущему методу.

```python
links.most_profitable(10)
```

```python
%timeit links.most_profitable(10)
```

#### longest()

Возвращает топ-n самых длинных по хронометражу фильмов. Длительность по времени на сайте представлена в часах и минутах, но после в скобках указано значение только в минутах. Его метод и берёт за основу, сохраняя в переменную, которая также будет добавлена в промежуточный вариант словаря. И именно по этим значениям будет происходить сортировка, однако в итоговый словарь попадаетл лишь полноценная исходная строка, как на сайте.

```python
links.longest(10)
```

```python
%timeit links.longest(10)
```

#### top_cost_per_minute()

Возвращает топ-n самых дорогих по стоимости за минуту фильмов. По сути, идентичен методу most_profitable(), но с отличием в том, что значениями для сортировки будут выступать результаты деления бюджета фильма на его продолжительность в минутах.

```python
links.top_cost_per_minute(10)
```

```python
%timeit links.top_cost_per_minute(10)
```

Также в классе Links имеются вспомогательные методы:
- file_reader() считывает поступающие на вход данные построчно, разделяя каждую строку на элементы, содержащие соответствующую информацию: "movieId", "imdbId" и "tmdbId". Получившийся список списков сохраняется в переменную класса для дальнейшего анализа.
- gen_read() реализует построчное считывание из файла с помощью генератора.
- check_id() проверяет наличие фильма в базе данных по "movieId" и в случае успеха возвращает либо "imdbId", либо "tmdbId" в зависимости от входных данных.
- request() делает запрос с помощью requests и возвращает ответ.
- parse() отвечает за всю логику получения данных с сайта IMDB. Здесь с помощью методов BeautifulSoup и регулярного выражения ищутся нужные поля и информация, соответствующая им. После чего найденные данные возвращаются для последующих манипуляций.
- skip() служит небольшой проверкой обрабатываемых данных на их пустоту или возвращённые ошибки из-за неудавшегося подключения. Чтобы не прерывать весь анализ из-за возможных случайных ошибок, данный метод позволяет игнорировать неудавшиеся попытки.
