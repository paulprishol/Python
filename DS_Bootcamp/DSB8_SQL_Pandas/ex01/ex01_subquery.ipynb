# Exercise 01. Subquery

```python
import pandas as pd
import sqlite3
```

## Create a connection to the database using the `sqlite3` library

```python
con = sqlite3.connect("../data/checking-logs.sqlite")
```

## Get the schema of the `checker` table

```python
query = "PRAGMA table_info(checker);"
df = pd.io.sql.read_sql(query, con)
df
```

## Get only the first ten rows of the table to see what the table looks like

```python
query = """
SELECT *
FROM checker
LIMIT 10
"""
df = pd.io.sql.read_sql(query, con)
df
```

## Count how many rows satisfy the following conditions using one query with any number of subqueries

```python
query = """
SELECT COUNT(*) AS cnt
FROM pageviews
WHERE uid IN (
    SELECT uid
    FROM checker
    WHERE status = 'ready'
    AND numTrials = 1
    AND labname IN ('laba04', 'laba04s',
    'laba05','laba06', 'laba06s', 'project1')
)
"""
checkers = pd.io.sql.read_sql(query, con)
checkers
```
