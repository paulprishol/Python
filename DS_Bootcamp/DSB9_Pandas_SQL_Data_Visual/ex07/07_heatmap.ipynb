# Exercise 07 : Heatmap

```python
import pandas as pd
import sqlite3
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
con = sqlite3.connect("../data/checking-logs.sqlite")
```

```python
days = {0: 'Mon', 1: 'Tue', 2: 'Wed', 3: 'Thu', 4: 'Fri', 5: 'Sat', 6: 'Sun'}
query = """
SELECT uid, timestamp
FROM checker
WHERE uid LIKE 'user_%'
"""
datetimes = pd.io.sql.read_sql(query, con, parse_dates=['timestamp'])
datetimes['weekday'] = datetimes['timestamp'].apply(lambda x: days[x.dayofweek])
datetimes['hour'] = datetimes['timestamp'].dt.hour
datetimes = datetimes.drop(columns=['timestamp'])
datetimes
```

```python
users = datetimes.groupby(['uid']).count().sort_values('hour', ascending=False).index
users
```

```python
weekdays = datetimes.groupby(['uid']).weekday.value_counts().unstack(0)
weekdays = weekdays.reindex(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'], columns=users).fillna(0)
weekdays
```

```python
index = pd.RangeIndex(stop=24)
hours = datetimes.groupby(['uid']).hour.value_counts().unstack(0)
hours = hours.reindex(index=index, fill_value=0, columns=users).rename_axis('hour').fillna(0)
hours
```

```python
fig, ax = plt.subplots(figsize=(15, 8))
im = ax.imshow(weekdays, cmap='RdBu')
divider = make_axes_locatable(ax)
cax = divider.append_axes('right', size='5%', pad=0.05)
plt.colorbar(im, cax=cax)
ax.set_xticks(range(weekdays.shape[1]))
ax.set_yticks(range(weekdays.shape[0]))
ax.set_xticklabels(weekdays.columns)
ax.set_yticklabels(weekdays.index)
plt.setp(ax.get_xticklabels(), rotation=90)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(15, 8))
im = ax.imshow(hours, cmap='jet')
divider = make_axes_locatable(ax)
cax = divider.append_axes('right', size='5%', pad=0.05)
plt.colorbar(im, cax=cax)
ax.set_xticks(range(hours.shape[1]))
ax.set_yticks(range(hours.shape[0]))
ax.set_xticklabels(hours.columns)
ax.set_yticklabels(hours.index)
plt.setp(ax.get_xticklabels(), rotation=90)
plt.show()
```
