# Exercise 05 : Boxplot

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../data/ab-test.csv')
df
```

```python
fig, ax = plt.subplots(1, 2, figsize=(15, 8), sharey=True)
df[df['time'] == 'after'].boxplot(column='diff',
                                  by='group',
                                  ax=ax[0],
                                  boxprops=dict(color='green', linewidth=3),
                                  medianprops=dict(linewidth=2),
                                  color=dict(whiskers='red', medians='red', caps='red'))
df[df['time'] == 'before'].boxplot(column='diff',
                                  by='group',
                                  ax=ax[1],
                                  boxprops=dict(color='green', linewidth=3),
                                  medianprops=dict(linewidth=2),
                                  color=dict(whiskers='red', medians='red', caps='red'))
ax[0].set_title('after')
ax[1].set_title('before')
fig.suptitle(
    'Lag before labs deadlines. After and before the Newsfeed', fontsize=15)
plt.show()
```

## What was the IQR of the control group before the newsfeed?

## The answer is __150 before__ and __80 after__, so we can see a significant reduction in the spread of values
