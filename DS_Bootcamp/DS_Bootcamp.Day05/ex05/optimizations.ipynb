## Iterations: in all the following subtasks, you need to calculate `fines/refund*year` for each row. Create a new column with the calculated data. Measure the time using the magic command `%%timeit` in the cell

```python
def using_loop(df):
    res = []
    for i in range(0, len(df)):
        res.append(df.iloc[i]['Fines'] / df.iloc[i]['Refund'] * df.iloc[i]['Year'])
    return res
df['Calculation'] = using_loop(df)
df
```

```python
%%timeit
df['Calculation'] = using_loop(df)
df
```

```python
def using_itterrows(df):
    res = []
    for _, row in df.iterrows():
        res.append(row['Fines'] / row['Refund'] * row['Year'])
    return res
df['Calculation'] = using_itterrows(df)
df
```

```python
%%timeit
df['Calculation'] = using_itterrows(df)
df
```

```python
df['Calculation'] = df.apply(func=lambda row: row['Fines'] / row['Refund'] * row['Year'] , axis='columns')
df
```

## Downcasting

```python
df.info(memory_usage='deep')
```

```python
optimized = df.copy()
optimized
```

```python
optimized['Fines'] = pd.to_numeric(optimized['Fines'], downcast='float')
optimized['Refund'] = pd.to_numeric(optimized['Refund'], downcast='integer')
optimized['Year'] = pd.to_numeric(optimized['Year'], downcast='integer')

optimized['Calculation'] = optimized['Calculation'].astype('float32')

optimized.info(memory_usage='deep')
```

## Categories

```python
optimized['Make'] = optimized['Make'].astype('category')
optimized['Model'] = optimized['Model'].astype('category')

optimized.info(memory_usage='deep')
```
