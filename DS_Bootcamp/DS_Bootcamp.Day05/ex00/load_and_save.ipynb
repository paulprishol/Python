# Exercise 00. Load and save

```python
import pandas as pd
```

## `read_csv`

```python
df = pd.read_csv('../data/feed-views.log', skiprows=[2, 3], skipfooter=2,
                  names=['datetime', 'user'], index_col='datetime', engine='python', sep='\t')
df.head()
```

```python
df.tail()
```

## rename `datetime` to `date_time`

```python
df.index.names = ['date_time']
df.head()
```

## `to_csv`

```python
filepath = '../data/feed-views-semicolon.log'
df.to_csv(filepath, sep=';')

with open(filepath, 'r') as file:
    for _ in range(5):
        print(file.readline().strip())
```
